```python
import glob
from skimage import io
# bounding box file and image files location
# Use r' ' in Windows, ' ' in Linux for folder location 
im_bbox = glob.glob(r'C:\Users\SAM8FH\Documents\Udacity\term3\yolo_stop\labels\stopsign_original\**\*.txt', recursive=True)
im_jpg = glob.glob(r'C:\Users\SAM8FH\Documents\Udacity\term3\yolo_stop\images\stopsign_jpeg\**\*.jpg', recursive=True)
print(im_bbox[3][75:])
print(im_jpg[3][71:])

for i in range(len(im_bbox)):
    print(im_bbox[i][75:], im_jpg[i][71:])  
    
```

```python
# test image 
# import cv2
im = io.imread(im_jpg[3])
print(im.shape)
```

```python
# parse all bounidng box text files
xml_list = []
for k in range(len(im_bbox)):
    f = open(im_bbox[k], encoding="utf-8")
    lines = f.readlines()
    for i in range(0, len(lines)):
        if i > 0:
            im = io.imread(im_jpg[k])
            im_name = im_jpg[k][71:]
            shape_tuple = tuple([im.shape[1], im.shape[0]])
            class_name = 'stop'
            x = tuple([int(i) for i in lines[i].split()])
            value = (im_name,) + shape_tuple + (class_name,) + tuple(x)
            xml_list.append(value)
#             print(x)
    
```

```python
print(len(xml_list))
print(xml_list[0])
```

```python
# make a dataframe of all bounding boxes
import pandas as pd
column_name = ['filename', 'width', 'height', 'class', 'xmin', 'ymin', 'xmax', 'ymax']
xml_df = pd.DataFrame(xml_list, columns=column_name)
```

```python
# write to csv 
xml_df.to_csv('yolo_stop_labels.csv', index=None)
```
