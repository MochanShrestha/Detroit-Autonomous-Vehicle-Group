```python
### Load the images and plot them here.
### Feel free to use as many code cells as needed.
import matplotlib.pyplot as plt
import cv2
import glob
# import tensorflow as tf
import numpy as np
from skimage import io

images_pos = glob.glob('stopsign_pos1_trimmed/**/*.jpg', recursive=True)
images_pos_array = []

for image in images_pos:
    images_pos_array.append(image)

print(len(images_pos_array))
```

```python
# pos_img_count = 0
# neg_img_count = 0

def classify_image(img, red_ratio_high_threshold = 0.1, red_ratio_low_threshold = 0.05 ):
    global pos_img_count, neg_img_count
#     global neg_img_count = 0
    r_channel = img[:,:,0]
    g_channel = img[:,:,1]
    b_channel = img[:,:,2]
    rgbbinary = np.zeros_like(r_channel)

    rgbbinary[(r_channel >= 85) & (g_channel <= 53 ) & (b_channel <= 53 )] = 1

    red_sum = np.ndarray.sum(rgbbinary) 
    total_pixels = rgbbinary.shape[0]*rgbbinary.shape[1]

    red_ratio = red_sum/total_pixels
    
#     print("total_pixels = ", total_pixels)
#     print("red_sum = ", red_sum)
    print("red_ratio = ", red_ratio)
    
    img_type = 0

#     plt.figure(figsize=(30,2))
#     plt.imshow(np.dstack(( rgbbinary*255, rgbbinary*0, rgbbinary*0)) )
#     red_ratio_threshold = 0.3

    # save_file_name = "sorted_pos/1.jpg"
    img_ext = ".jpg"
    folder_pos = "sorted_pos5/"
    folder_neg = "sorted_neg5/"
    
#     img_resized = cv2.resize(img, (64, 60))
    img_resized = img
    img_resized_BGR = img_resized[...,::-1]
    
    if red_ratio >= red_ratio_high_threshold and red_ratio < 0.3:
        
        pos_img_count = pos_img_count + 1
    
        save_file_name = "".join([folder_pos, str(pos_img_count), img_ext])
        cv2.imwrite(save_file_name, img_resized_BGR)
        img_type = 1
    elif red_ratio <= red_ratio_low_threshold:
        
        folder_ = "sorted_pos/0"
        neg_img_count = neg_img_count + 1
    
        save_file_name = "".join([folder_neg, str(neg_img_count), img_ext])
       
        cv2.imwrite(save_file_name, img_resized_BGR)
        img_type = -1
        
#     print(save_file_name)

    # ''.join([a, b]) better than + for cpmutational efficiency O(n) vs O(n^2)
    # cv2.cvtColor(out_img, cv2.COLOR_RGB2HLS)
    return img_type


def split_data(image, img_type = 1):
    global pos_img_count, neg_img_count    
    nrow = image.shape[1]
    ncol = image.shape[0]

    num_rsplit = 5
    num_csplit = 4
    r_ovrlap = 0.5
    c_ovrlap = 0.5
    
    row_width = int(nrow / num_rsplit)
    col_width = int(ncol / num_csplit)
    
#     print(row_width, col_width)
    
#     row_start = 120
#     row_end = row_start + 80 #row_width
    
    
#     col_start = 80
#     col_end = col_start + 80 # col_width
#     = 5 + 5*1 -1 
    num_r_total = int(num_rsplit + num_rsplit * (1/(1 - r_ovrlap) - 1)) - 1
    num_c_total = int(num_csplit + num_csplit * (1/(1 - c_ovrlap) - 1)) - 1
    
    for i in range(num_r_total):
        for j in range(num_c_total):
            row_start = i*int(row_width*(1 - r_ovrlap))
            row_end = row_start + row_width
            
            col_start = j*int(col_width*(1 - c_ovrlap))
            col_end = col_start + col_width
            crop_img = image[col_start:col_end, row_start:row_end,] 
            
            image_type = classify_image(crop_img, 0.12, 0.04)
#             print("Image type: ", image_type)         
    
            
    
    
    
#     crop_img = image[col_start:col_end, row_start:row_end,] 
    
    
#     return image_new
    return pos_img_count, neg_img_count



# split_status = split_data(im7)
# print("pos_count & neg_count: ", split_status)


```
