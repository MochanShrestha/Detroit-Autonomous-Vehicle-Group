```python
### Load the images and plot them here.
### Feel free to use as many code cells as needed.
import matplotlib.pyplot as plt
import cv2
import glob
%matplotlib inline

# import tensorflow as tf
import numpy as np
from skimage import io

images_pos = glob.glob('/home/mahesh/davg/StopSignData/stopsign_pos1/**/*.jpg', recursive=True)
images_pos_array = []

for image in images_pos:
    images_pos_array.append(image)

print(len(images_pos_array))
```

```python
def filter_red(img):
    r_channel = img[:,:,0]
    g_channel = img[:,:,1]
    b_channel = img[:,:,2]
    rgbbinary = np.zeros_like(r_channel)

    rgbbinary[(r_channel >= 85) & (g_channel <= 48 ) & (b_channel <= 64 )] = 1

    return rgbbinary
    
im = io.imread(images_pos_array[1])
rgbbinary = filter_red(im)  
plt.figure(figsize=(30,2))
plt.imshow(np.dstack(( rgbbinary*255, rgbbinary*0, rgbbinary*0)) )
```

```python

num_files = 5
im_array = []

for i in range(num_files):
    im_array.append(io.imread(images_pos_array[i]))

fig = plt.figure(figsize=(30,3))
for i in range(num_files):
    im = io.imread(images_pos_array[i])
    rgbbinary = filter_red(im) 
    a = fig.add_subplot(1, num_files, i+1)
    plt.imshow(im)
    
```
