```python
### Load the images and plot them here.
### Feel free to use as many code cells as needed.
import matplotlib.pyplot as plt
import cv2
import glob
%matplotlib inline

# import tensorflow as tf
import numpy as np
from skimage import io

images_pos = glob.glob('/home/mahesh/davg/StopSignData/stopsign_pos1/**/*.jpg', recursive=True)
images_pos_array = []

for image in images_pos:
    images_pos_array.append(image)

print(len(images_pos_array))
```

```python
def filter_red(img):
    r_channel = img[:,:,0]
    g_channel = img[:,:,1]
    b_channel = img[:,:,2]
    rgbbinary = np.zeros_like(r_channel)

    rgbbinary[(r_channel >= 85) & (g_channel <= 48 ) & (b_channel <= 64 )] = 1

    return rgbbinary
    
im = io.imread(images_pos_array[1])
rgbbinary = filter_red(im)  
im_red = np.dstack(( rgbbinary*255, rgbbinary*0, rgbbinary*0))
plt.figure(figsize=(30,2))
plt.imshow(im_red )
```

```python
# https://github.com/nikgens/TankRobotProject/blob/master/signRecognition/signRecognition.py
from imutils.perspective import four_point_transform
from imutils import contours
import imutils
gray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
blurred = cv2.GaussianBlur(gray, (5, 5), 0)
edged = cv2.Canny(blurred, 50, 200, 255)

# find contours in the edge map, then sort them by their
# size in descending order
cnts = cv2.findContours(edged.copy(), cv2.RETR_EXTERNAL,
        cv2.CHAIN_APPROX_SIMPLE)
# The is_cv2() and is_cv3() are simple functions that can be used to 
# automatically determine the OpenCV version of the current environment
# cnts[0] or cnts[1] hold contours
cnts = cnts[0] if imutils.is_cv2() else cnts[1]
cnts = sorted(cnts, key=cv2.contourArea, reverse=True)
displayCnt = None
```
